## Q5) Using loc and transform to create loyalty flag



```python
# Using loc and transform to create loyalty flag
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# View updated data frame
ords_prods_merge.head(20)
```

```python
# Use loc function to make loyalty flag column with given criteria
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check frequency of values in new loyalty flag column
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# View updated data frame
ords_prods_merge.head()
```

## Q6) Different types of spenders in their marketing campaigns

```python
# Using loc and transform to create loyalty flag
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# View updated data frame
ords_prods_merge.head(20)
```

```python
#Use loc function to make spending flag column with given criteria
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] <10, 'spender_type'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >=10, 'spender_type'] = 'High spender'
```

```python
# Check frequency of values in new spending flag column
ords_prods_merge['spender_type'].value_counts(dropna = False)
```

```python
# View relevant columns of updated data frame
ords_prods_merge[['user_id', 'mean_product_price', 'spender_type']].head(20)
```

## Q7) Determine frequent versus non-frequent customers

```python
# Using loc and transform to create frequency flag
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)

```

```python
# View updated data frame
ords_prods_merge.head(20)

```

```python
#Use loc function to make order frequency flag column with given criteria
ords_prods_merge.loc[ords_prods_merge['order_frequency'] >=20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['order_frequency']>10) & (ords_prods_merge['order_frequency'] <=20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <=10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check frequency of values in new spending flag column
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
